# 4. Write a Python program to find the volume of a sphere with diameter 12 cm.
# Formula: V=4/3 *  * r^3

```python
def vol(rad):
    return ((4/3)*3.142*rad*rad*rad)
diameter = 12
radius = diameter / 2
print(vol(radius))
```

# 2. Create the below pattern using nested for loop in Python.
# *
# * *
# * * *
# * * * *
# * * * * *
# * * * *
# * * *
# * *
# *

```python
for a in range(5):
    for b in range(a+1):
        print('*',end = ' ')
    print('\n')
for c in range(5,1,-1):
    for d in range(c-1,0,-1):
        print('*',end = ' ')
    print('\n')
```
