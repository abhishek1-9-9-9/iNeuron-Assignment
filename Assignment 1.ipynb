# 4. Write a Python program to find the volume of a sphere with diameter 12 cm.
# Formula: V=4/3 * π * r^3

```python
def volume(radius):
    return ((4/3)*3.142*radius*radius*radius)
diameter = 12
rad = diameter / 2
print(volume(rad))
```

# 2. Create the below pattern using nested for loop in Python.
# *
# * *
# * * *
# * * * *
# * * * * *
# * * * *
# * * *
# * *
# *

```python
for row1 in range(5):
    for col1 in range(row1+1):
        print('*',end = ' ')
    print('\n')
for row2 in range(5,1,-1):
    for col2 in range(row2-1,0,-1):
        print('*',end = ' ')
    print('\n')
```
